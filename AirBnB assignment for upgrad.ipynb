### Task 4

You want to observe the relationship between the reviews(given by `number_of_reviews`) and the ratings received (`review_scores_rating`) by different AirBnB listings.

For this, plot both a scatterplot and a jointplot. What can you say about the relationship between ratings and reviews?

```python
#Here we are using the Scatter plot as mentioned above,on x-axis we apply the reviews and on Y-axis we apply the ratings but additionally we are adding one hue which is room_type
sns.scatterplot(data=df,
                x="number_of_reviews",
                y="review_scores_rating",
                hue="room_type")
plt.title("Relationship between Review and Ratings")
plt.show()
```

```python
#same as above but joint plot is used to plot to graph and show the relation.
sns.jointplot(data=df,x="number_of_reviews",y="review_scores_rating",hue="room_type")
plt.title("Distribution of number of ratings")
plt.xlabel("number of ratings")
plt.ylabel("Frequency")
plt.show()
```

**Observation** - Here most of the guest has choosen the high rated and highly review AirBnB And the most Room type choosen by the guests are the Private Room and after that Entire home/Appratment 

### Task 5

### Task 6

You wish to identify the relationship between the various ratings each of the listing has received from the customers.
These ratings have been summarized below:

```python
review_columns = ['review_scores_rating',
 'review_scores_accuracy',
 'review_scores_cleanliness',
 'review_scores_checkin',
 'review_scores_communication',
 'review_scores_location',
 'review_scores_value']
review_columns
```

```python
#for creating Heatmap We need correlaton . here review_columns is given in question
df[review_columns].corr()
```

```python
#Creating the Heatmap 
sns.heatmap(df[review_columns].corr(), cmap="Oranges")
plt.title("Relationship between the various ratings")
plt.xlabel("Various ratings")
plt.show()
```

**Observation** - Darker shades represent stronger positive correlations and the lighter shades represent weaker or negative correlations. example, positive correlations between 'cleanliness' and 'accuracy' could suggest that guests choose the accuracy higher when they see the property to be clean.

### Task 7 

Analyze the listings' prices across the following neighborhoods using a categorical boxplot
- 'Westerpark', 'Oud-Noord', 'Noord-West','Zuid'

```python
#here we create a variable called neighbour in Dataframe
neighbour = ['Westerpark', 'Oud-Noord', 'Noord-West','Zuid']
```

```python


df2 = df.loc[df['neighbourhood'].isin(neighbour)]
df2.shape

```

```python
#creating the box plot
sns.boxplot(x="neighbourhood", y="price", data=df2)
plt.title("price according to the neighbour")
plt.show()
```

**Observation** - The plot shows the variation in prices among given neighbourhood, providing insights into the disparity in accommodation costs across Amsterdam.
By comparing the median and distribution of prices across the given neighbourhood, we can find which areas that are generally more affordable or luxurious.
